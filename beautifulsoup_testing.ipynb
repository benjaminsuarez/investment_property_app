```python
#url = "https://www.daft.ie/dublin-city/residential-property-for-rent/?s%5Bmxp%5D=700&s%5Bsort_type%5D=a&s%5Bsort_by%5D=price&s%5Bignored_agents%5D%5B0%5D=1551"
url = "https://www.daft.ie/dublin-city/residential-property-for-rent/?s%5Bmxp%5D=750&s%5Bsort_type%5D=a&s%5Bsort_by%5D=price&s%5Bignored_agents%5D%5B0%5D=1551"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0"} #need header or wont work
r = requests.get(url,headers=headers)
```
